```python
import requests

# Fetch a web page
r = requests.get("https://www.investopedia.com/vanguard-finds-only-top-earners-are-financially-on-track-for-retirement-8348641")
print(r.text)
```

```python
import re

# Remove HTML tags using RegularExpression
pattern = re.compile(r'<.*?>')  # tags look like <...>
print(pattern.sub('', r.text))  # replace them with blank
```

```python
from bs4 import BeautifulSoup
```

```python
# Remove HTML tags using Beautiful Soup library
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python
#mw-content-container
# Find all articles
summaries = soup.find_all("div", class_="comp mntl-sc-block-callout-body mntl-text-block")
summaries[0]
```

```python
summaries[0].get_text().strip()
```
