```python
import numpy as np
import pandas as pd
#visualize
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

pokemon=pd.read_csv('/content/pokemon.csv')






```

```python
print(pokemon.shape)
```

```python
print(pokemon.head)
```

```python
pokemon.dropna(axis=0,inplace=True)
pokemon.isnull().sum()
```

```python
pokemon['generation_id'].value_counts()
```

```python
sb.countplot(data=pokemon,x='generation_id');
```

```python
sb.color_palette()
```

```python
base_color=sb.color_palette()[6]
```

```python
sb.countplot(data=pokemon,x='generation_id',color=base_color);
```

```python
pokemon['generation_id'].value_counts()
```

```python
gen_order=pokemon['generation_id'].value_counts().index
```

```python
sb.countplot(data=pokemon,x='generation_id',color=base_color,order=gen_order);
```

```python
pokemon.info()
```

```python
pokemon['type_1'].value_counts()
```

```python
type_order=pokemon['type_1'].value_counts().index
```

```python
sb.countplot(data=pokemon,x='type_1',color=base_color,order=type_order);
```

```python
sb.countplot(data=pokemon,y='type_1',color=base_color,order=type_order);
```

```python
sorted_counts = pokemon['generation_id'].value_counts()
plt.pie(sorted_counts,labels=sorted_counts.index,startangle=90,counterclock=False);
plt.axis('square');
```
